## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Marj's work :)
# check for the datatypes for each column
schools_students_df.dtypes
```

```python
# Marj's work :)
# Calculate the total number of schools
group_schools = schools_students_df.groupby("school_name")
total_schools = len(group_schools)

# Calculate the total number of students
total_students = len(schools_students_df["Student ID"])

# Calculate the total budget 
total_budget = schools_df["budget"].sum()


# Calculate the average math score
avg_math_score = schools_students_df["math_score"].mean()

# Calculate the average reading score
avg_reading_score = schools_students_df["reading_score"].mean()

# Calculate the percentage of students with a passing math score (70 or greater)
math_pass = len(schools_students_df.loc[schools_students_df["math_score"] >= 70]) 
per_math = math_pass / total_students * 100

# Calculate the percentage of students with a passing reading score (70 or greater)
reading_pass = len(schools_students_df.loc[schools_students_df["reading_score"] >= 70]) 
per_reading = reading_pass/ total_students * 100

# Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_passing = schools_students_df[(schools_students_df["math_score"] >= 70) &
                                      (schools_students_df["reading_score"] >= 70)].count()["student_name"]
                
overall_pass_per = overall_passing/total_students * 100
    

# Create a dataframe to hold the above results
district_sum_df = pd.DataFrame({
    "Total Schools":[total_schools],
    "Total Students":[total_students],
    "Total Budget":[total_budget],
    "Average Math Score": [avg_math_score],
    "Average Reading Score": [avg_reading_score],
    "% Passing Math": [per_math],
    "% Passing Reading": [per_reading],
    "% Overall Passing": [overall_pass_per]
})

# Optional: give the displayed data cleaner formatting
district_sum_df["Total Students"] = district_sum_df["Total Students"].map("{:,}".format)
district_sum_df["Total Budget"] = district_sum_df["Total Budget"].map("${:,.2f}".format)
district_sum_df["Average Math Score"] = district_sum_df["Average Math Score"].map("{:,.2f}".format)
district_sum_df["Average Reading Score"] = district_sum_df["Average Reading Score"].map("{:,.2f}".format)
district_sum_df["% Passing Math"] = district_sum_df["% Passing Math"].map("{:,.2f}%".format)
district_sum_df["% Passing Reading"] = district_sum_df["% Passing Reading"].map("{:,.2f}%".format)
district_sum_df["% Overall Passing"] = district_sum_df["% Overall Passing"].map("{:,.2f}%".format)

# adding a title in case anything is unclear
title = "DISTRICT SUMMARY"
print(title)

# output for the summary table as defined above
district_sum_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Marj's work :)

# Create an overview table that summarizes key metrics about each school, including:
# School Name, School Type, Total Students, Total School Budget, Per Student Budget,
# Average Math Score, Average Reading Score, % Passing Math, % Passing Reading,
# % Overall Passing (The percentage of students that passed math and reading.)
# Create a dataframe to hold the above results


# check for the school type
school_type = schools_df.set_index(["school_name"])["type"]

# count all students
students_per_school = schools_students_df["school_name"].value_counts()

# calculate the total and per student budgets
budget_per_school = schools_students_df.groupby(["school_name"]).mean()["budget"]
budget_per_student= budget_per_school/students_per_school

# calculate the average math score per school
math_per_school = schools_students_df.groupby(["school_name"]).mean()["math_score"]

# calculate the average reading score per school
read_per_school = schools_students_df.groupby(["school_name"]).mean()["reading_score"]

# calculate the # and % of students with a passing math score (70 or greater) per school
math_pass_school = schools_students_df[(schools_students_df["math_score"] >= 70)]
per_math_school = math_pass_school.groupby(["school_name"]).count()["student_name"] / students_per_school *100

# calculate the # and % of students with a passing reading score (70 or greater) per school
read_pass_school = schools_students_df[(schools_students_df["reading_score"] >= 70)]
per_read_school = read_pass_school.groupby(["school_name"]).count()["student_name"] / students_per_school *100

# calculate the # and percent of students who passed math and reading per school
math_read_school = schools_students_df[(schools_students_df["math_score"] >= 70) &
                                      (schools_students_df["reading_score"] >= 70)]
per_overall_school = (math_read_school.groupby(["school_name"]).count()["student_name"] / students_per_school) *100

# the dataframe and output!
school_summary_df = pd.DataFrame({
    "School Type":school_type,
    "Total Students":students_per_school,
    "Total School Budget":budget_per_school, 
    "Per Student Budget": budget_per_student,
    "Average Math Score": math_per_school,
    "Average Reading Score": read_per_school,
    "% Passing Math": per_math_school,
    "% Passing Reading": per_read_school,
    "% Overall Passing": per_overall_school
})

# making the output pretty
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.2f}".format)
school_summary_df["Average Math Score"] = school_summary_df["Average Math Score"].map("{:.2f}".format)
school_summary_df["Average Reading Score"] = school_summary_df["Average Reading Score"].map("{:.2f}".format)
school_summary_df["% Passing Math"] = school_summary_df["% Passing Math"].map("{:.2f}%".format)
school_summary_df["% Passing Reading"] = school_summary_df["% Passing Reading"].map("{:.2f}%".format)
school_summary_df["% Overall Passing"] = school_summary_df["% Overall Passing"].map("{:.2f}%".format)

# adding a title in case anything is unclear
title2 = "SCHOOL SUMMARY"
print(title2)

# display the output
school_summary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Marj's work :)

# series for the 9th-12th graders
grd_nine = schools_students_df[(schools_students_df["grade"] == "9th")]
grd_ten = schools_students_df[(schools_students_df["grade"] == "10th")]
grd_eleven = schools_students_df[(schools_students_df["grade"] == "11th")]
grd_twelve = schools_students_df[(schools_students_df["grade"] == "12th")]

# now group by school for math scores
grd_nine_math = grd_nine.groupby(["school_name"]).mean()["math_score"]
grd_ten_math = grd_ten.groupby(["school_name"]).mean()["math_score"]
grd_eleven_math = grd_eleven.groupby(["school_name"]).mean()["math_score"]
grd_twelve_math = grd_twelve.groupby(["school_name"]).mean()["math_score"]

# bring both series into a new DataFrame
math_grade_scores_df = pd.DataFrame({
    "9th":grd_nine_math,
    "10th":grd_ten_math,
    "11th":grd_eleven_math, 
    "12th": grd_twelve_math
})

# making it pretty
math_grade_scores_df["9th"] = math_grade_scores_df["9th"].map("{:.2f}".format)
math_grade_scores_df["10th"] = math_grade_scores_df["10th"].map("{:.2f}".format)
math_grade_scores_df["11th"] = math_grade_scores_df["11th"].map("{:.2f}".format)
math_grade_scores_df["12th"] = math_grade_scores_df["12th"].map("{:.2f}".format)

# adding a title in case anything is unclear
title5 = "MATH SCORE AVERAGES PER SCHOOL, BY GRADE"
print(title5)

math_grade_scores_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Marj's work :)

# to approximate the bin ranges, check the max and min of the budget per students. We have a range from $578-$655
# setting the ranges in groups of $25: 575-599, 600-624, 625-649, 650-674

# let's create the bins
school_spending_bins = [575, 600, 625, 650, 675]
spending_bin_names = ["$575-$599", "$600-$624","$625-$649", "$650-$674"]

# copy the per student budget from the school summary above to have flexibility
school_spending_df = school_summary_df
school_spending_df.head(2)
```

```python
# Marj's work :)

# throwing the data into our bins
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(budget_per_student, school_spending_bins, labels=spending_bin_names)
school_spending_df
```

```python
# checking the school spending types

school_spending_df.dtypes
```

```python
# Marj's work :)

#moroe copies for flexibility
school_spending_copy = school_spending_df
school_spending_copy.head(3)
```

```python
# Marj's work :)

school_spending_copy.dtypes
```

```python
# Marj's work :)

# update the data types so the math can be done!

school_spending_copy["Average Math Score"] = school_spending_df["Average Math Score"].astype('float')
school_spending_copy["Average Reading Score"] =  school_spending_df["Average Reading Score"].astype('float')
school_spending_copy["% Passing Reading"] =  school_spending_df["% Passing Reading"].str.replace("%","").astype('float')
school_spending_copy["% Passing Math"] =  school_spending_df["% Passing Math"].str.replace("%","").astype('float')
school_spending_copy["% Overall Passing"] =  school_spending_df["% Overall Passing"].str.replace("%","").astype('float')
school_spending_copy.dtypes

school_spending_copy.head(3)
```

```python
# Marj's work :)

# DO THE MATH... 
# Average Math Score
spending_scores_math_avg = school_spending_copy.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]

# Average Reading Score
spending_scores_read_avg = school_spending_copy.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]

# % Passing Math
spending_scores_math_per = school_spending_copy.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]

# % Passing Reading
spending_scores_read_per = school_spending_copy.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]

# Overall Passing Rate (Average of the above two)
spending_scores_both_per = school_spending_copy.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

```

```python
# Marj's work :)

# throw it in a DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score": spending_scores_math_avg.round(2),
    "Average Reading Score": spending_scores_read_avg.round(2),
    "% Passing Math":spending_scores_math_per.round(2),
    "% Passing Reading":spending_scores_read_per.round(2),
    "% Overall Passing":spending_scores_both_per.round(2)
})

# adding a title in case anything is unclear
title6 = "SCORES BY SCHOOL SPENDING"
print(title6)

spending_summary
```
